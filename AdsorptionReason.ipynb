```python
import csv
from collections import defaultdict

def group_by_material_name_with_names(csv_file_path):
    """
    æ ¹æ® CSV æ–‡ä»¶ä¸­çš„ 'Materials Name' åˆ—å¯¹æ•°æ®è¿›è¡Œåˆ†ç»„ï¼Œ
    å¹¶å°†ç›¸åŒ 'Materials Name' çš„è¡Œæ•°æ®åˆå¹¶æˆä¸€ä¸ªå­—ç¬¦ä¸²ï¼Œ
    åŒæ—¶è¿”å›ä¸€ä¸ªåŒ…å«å¯¹åº”ææ–™åç§°çš„åˆ—è¡¨ã€‚

    Args:
        csv_file_path (str): CSV æ–‡ä»¶çš„è·¯å¾„ã€‚

    Returns:
        tuple: ä¸€ä¸ªåŒ…å«ä¸¤ä¸ªå…ƒç´ çš„å…ƒç»„ã€‚
               - ç¬¬ä¸€ä¸ªå…ƒç´ æ˜¯ä¸€ä¸ªåˆ—è¡¨ï¼Œå…¶ä¸­æ¯ä¸ªå…ƒç´ æ˜¯ä¸€ä¸ªå­—ç¬¦ä¸²ï¼ŒåŒ…å«äº†å…·æœ‰ç›¸åŒ 'Materials Name' çš„æ‰€æœ‰è¡Œã€‚
               - ç¬¬äºŒä¸ªå…ƒç´ æ˜¯ä¸€ä¸ªåˆ—è¡¨ï¼ŒåŒ…å«äº†ä¸ç¬¬ä¸€ä¸ªåˆ—è¡¨ä¸­çš„å­—ç¬¦ä¸²ä¸€ä¸€å¯¹åº”çš„ææ–™åç§°ã€‚
    """
    grouped_data = defaultdict(list)
    material_names_list = []

    with open(csv_file_path, 'r', newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        header = next(reader)  # è¯»å–æ ‡é¢˜è¡Œ
        for row in reader:
            material_name = row[0]
            grouped_data[material_name].append(row)

    result_strings = []
    
    result_names = []
    for material, rows in grouped_data.items():
        combined_string = ""
        combined_string += ",".join(header) + "\n" # å…ˆåŠ ä¸Šæ ‡é¢˜è¡Œ
        for row in rows:
            combined_string += ",".join(row) + "\n"
        result_strings.append(combined_string.strip())
        result_names.append(material)

    return result_strings, result_names

# å‡è®¾ä½ çš„ CSV æ–‡ä»¶åä¸º 'your_data.csv'
csv_file = 'output_to_llm.csv'
grouped_strings, material_names = group_by_material_name_with_names(csv_file)

# æ‰“å°ç»“æœ
for i, s in enumerate(grouped_strings):
    print(f"ææ–™åç§°: {material_names[i]}")
    print(s)
    print("-" * 20)

print("\nææ–™åç§°åˆ—è¡¨:", material_names)
result = {}
```
